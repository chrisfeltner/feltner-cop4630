##1a) A local minimum at (0, 0, 0) - 2D contour plot

```
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes()
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)

def f(x, y):
  return x*x + y*y

x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 50)
X, Y = np.meshgrid(x, y)

Z = f(X, Y)
plt.contour(X, Y, Z, colors='black')
plt.scatter(0, 0, color='red')
plt.show()
```

##1b) A local minimum at (0, 0, 0) - 3D Contour plot

```
ax = plt.axes(projection='3d')
ax.set_xlim3d(-20, 20)

ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.scatter(0, 0, 0, s=300, color='red')
```

##1c) A saddle point at (0, 0, 0) - 2D contour plot

```
ax = plt.axes()
def g(x, y): 
  return x*x*x+y*y*y

Z = g(X, Y)
plt.contour(X, Y, Z, colors='black')
plt.scatter(0, 0, c='red')
plt.show()
```

##1d) A saddle point at (0, 0, 0) - 3D contour plot

```
ax = plt.axes(projection='3d')

ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.scatter(0, 0, 0, s=500, color='red')
plt.show()
```

## 1e) A local maximum at (pi/2, pi/2) - 2D contour plot

```
ax = plt.axes()

def h(x, y):
  return np.sin(x) + np.sin(y)

Z = h(X, Y)
plt.contour(X, Y, Z, colors='black')
plt.scatter(np.pi/2, np.pi/2, color='red')
```

## 1f) A local maximum at (pi/2, pi/2, 2) - 3D Contour plot

```
ax = plt.axes(projection='3d')

ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.scatter(np.pi/2, np.pi/2, np.sin(np.pi/2) + np.sin(np.pi/2), s=100, color='red')
plt.show()
```
